```python
# http://shatterline.com/blog/2013/09/12/not-so-naive-classification-with-the-naive-bayes-classifier/
# https://chrisalbon.com/machine-learning/naive_bayes_classifier_from_scratch.html
import pandas as pd
data = pd.read_csv('http://www.shatterline.com/MachineLearning/data/tennis_anyone.csv')
data.loc[data['Play-Tennis'].isin(['Yes', 'No'])]
```

```python
total_entity = data.shape[0]
yes_total = data['Play-Tennis'][data['Play-Tennis'] == 'Yes'].count()
no_total = data['Play-Tennis'][data['Play-Tennis'] == 'No'].count()
```

```python
# Outlook 
sunny_yes = len(data.loc[(data['Outlook'] == 'Sunny') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
sunny_no = len(data.loc[(data['Outlook'] == 'Sunny') & (data['Play-Tennis'] == 'No')]) / no_total     

rain_yes = len(data.loc[(data['Outlook'] == 'Rain') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
rain_no = len(data.loc[(data['Outlook'] == 'Rain') & (data['Play-Tennis'] == 'No')]) / no_total     

overcast_yes = len(data.loc[(data['Outlook'] == 'Overcast') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
overcast_no = len(data.loc[(data['Outlook'] == 'Overcast') & (data['Play-Tennis'] == 'No')]) / no_total     
```

```python
# Temperature
hot_yes = len(data.loc[(data['Temperature'] == 'Hot') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
hot_no = len(data.loc[(data['Temperature'] == 'Hot') & (data['Play-Tennis'] == 'No')]) / no_total     

mild_yes = len(data.loc[(data['Temperature'] == 'Mild') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
mild_no = len(data.loc[(data['Temperature'] == 'Mild') & (data['Play-Tennis'] == 'No')]) / no_total     

cool_yes = len(data.loc[(data['Temperature'] == 'Cool') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
cool_no = len(data.loc[(data['Temperature'] == 'Cool') & (data['Play-Tennis'] == 'No')]) / no_total     
```

```python
# Humidity
high_yes = len(data.loc[(data['Humidity'] == 'High') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
high_no = len(data.loc[(data['Humidity'] == 'High') & (data['Play-Tennis'] == 'No')]) / no_total     

normal_yes = len(data.loc[(data['Humidity'] == 'Normal') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
normal_no = len(data.loc[(data['Humidity'] == 'Normal') & (data['Play-Tennis'] == 'No')]) / no_total     
```

```python
# wind
weak_yes = len(data.loc[(data['Wind'] == 'Weak') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
weak_no = len(data.loc[(data['Wind'] == 'Weak') & (data['Play-Tennis'] == 'No')]) / no_total     

strong_yes = len(data.loc[(data['Wind'] == 'Strong') & (data['Play-Tennis'] == 'Yes')]) / yes_total          
strong_no = len(data.loc[(data['Wind'] == 'Strong') & (data['Play-Tennis'] == 'No')]) / no_total     
```

```python
# question:
# Will players play tennis if whether is [Rain, Mild, High, Weak]?

# Naive Bayes Classifier
# P(class|feature) = P(feature|class)*P(class)/P(feature)

# P(Yes|features) = P(Outlook=Rain|Yes)*P(Temperature=Mild|Yes)*P(Humidity=High|Yes)*P(Wind=Weak)*P(Yes)                
# P(no|features) = P(Outlook=Rain|no)*P(Temperature=Mild|no)*P(Humidity=High|no)*P(Wind=Weak)*P(no)                

yes = rain_yes*mild_yes*high_yes*weak_yes*yes_total
no = rain_no*mild_no*high_no*weak_no*no_total

y = yes/(yes+no)
n = no/(yes+no)

answer = 'YES' if y > n else 'NO'

print('Output is: ', answer)
```
